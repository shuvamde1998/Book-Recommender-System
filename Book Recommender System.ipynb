```python
# Nearest Neighbour Algorithm(Brute algorithm) As there are many rows and many columns we cant plot graph
```

```python
from scipy.sparse import csr_matrix
book_matrix=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_matrix)
```

```python
book_pivot.index
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=5)
```

```python
def recommender(bookname):
    b_id=np.where(book_pivot.index==bookname)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[b_id,:].values.reshape(1,-1),n_neighbors=5)
    for i in suggestion[0]:
        print(book_pivot.index[i])
```

```python
recommender('Harry Potter and the Chamber of Secrets (Book 2)')
```
